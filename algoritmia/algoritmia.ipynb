### **CALCULA LA LONGITUD**
### **DE LA RACHA MÁS LARGA** 
### **DE NÚMEROS CONSECUTIVOS** 

```python
numeros = [1,2,3,4,4,5,6,7,7,7,8,9,9,9,9,10]

def racha_mas_larga(numeros : list) -> int:
    racha_actual = 1 
    racha_maxima = 1

    for indice in range(1, len(numeros)):
        if numeros[indice] == numeros[indice -1]:
            racha_actual+=1
        else:
            racha_actual = 1
        
        racha_maxima = max(racha_actual, racha_maxima)
    
    return racha_maxima if numeros else 0

print(f"La racha de numeros consecutivos más larga en esta lista es de: {racha_mas_larga(numeros)}")
```
