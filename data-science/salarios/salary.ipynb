### **¿CUÁNTO GANA UN**
### **DATA ENGINEER?**

```python
import pandas as pd

df = pd.read_csv("glassdoor_jobs_clean.csv")
de = df[df["Job Title"].str.contains("Data Engineer", case=False, na=False)].copy()

parts = de["Salary Estimate"].str.extract(r"\$?(\d+)[Kk]\s*-\s*\$?(\d+)[Kk]")
de["salary_num"] = (pd.to_numeric(parts[0], errors="coerce") + pd.to_numeric(parts[1], errors="coerce")) / 2

print(f"El salario promedio de un Data Engineer es {de['salary_num'].mean():.0f}K $/año.")
```
