### **¿CUÁNTO GANA UN**
### **DATA ENGINEER?**

```python
import pandas as pd

df = pd.read_csv("glassdoor_jobs_clean.csv")

# 1. Filtramos los trabajos de Data Engineer
data_engineers = df[df["Job Title"].str \
        .contains("Data Engineer", case=False, na=False)].copy()

# 2. Extraemos el rango del salario estimado parseando el string
salary_range = data_engineers["Salary Estimate"].str \
        .extract(r"\$?(\d+)[Kk]\s*-\s*\$?(\d+)[Kk]")

# 3. Calculamos la media del rango del salario estimado
salary_range["salary_num"] = (pd.to_numeric(salary_range[0])
                     + pd.to_numeric(salary_range[1])) / 2


print(f"El salario promedio es {salary_range['salary_num'].mean():.0f}K")
```
