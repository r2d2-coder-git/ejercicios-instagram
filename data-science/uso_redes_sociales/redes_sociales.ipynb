### ¿CUÁNTAS HORAS 
### CONSUME UN USUARIO 
### AL DÍA DE REDES SOCIALES?

```python
import pandas as pd

# Cargar datos
df = pd.read_csv('social_media_mental_health.csv')

# 1. Sumar horas diarias por usuario (por si hay múltiples filas por User_ID)
user_daily_hours = (
    df.groupby("User_ID", as_index=False)["Daily_Screen_Time_Hours"]
      .sum()
)

# 2. Calcular la media de horas diarias entre usuarios (average user)
average_user_hours = user_daily_hours["Daily_Screen_Time_Hours"].mean()

print(f"Un usuario promedio pasa {average_user_hours:.2f} horas al día en redes sociales.")
```

### ¿CON QUÉ EDAD
### SE CONSUME MÁS LAS REDES SOCIALES?

```python
import pandas as pd

df = pd.read_csv("social_media_mental_health.csv")
result = df.groupby("Age")["Daily_Screen_Time_Hours"].mean()
print(f"Edad: {result.idxmax()}, Horas de media: {result.max().round(2)}")
```
